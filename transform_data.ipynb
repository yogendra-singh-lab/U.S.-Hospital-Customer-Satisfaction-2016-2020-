### since `hcahps_linear_mean_value` and `patient_survey_star_rating` are representations of the boxed responses the two columns will be kept for this analysis

`Not Applicable` entries in these two columns will be represented with `np.nan` and `Not Available` with `0`,

then only rows with values greater than or equal to `0` values in both columns will be kept. This would take out all __boxed answer entries__,

a column called master_metric will be made. This column contains entries from both columns. Remember that any row entry that has a valid `patient_survey_star_rating` entry will have a `Not Applicable` entry (now `np.nan`) in `hcahps_linear_mean_value` and vice-versa. So both columns can be summed rowwise and no data will be lost

````python
data[["patient_survey_star_rating", "hcahps_linear_mean_value"]] = \
    (data[["patient_survey_star_rating", "hcahps_linear_mean_value"]]
        .applymap(lambda x: np.nan if x == "Not Applicable" else (0 if x == "Not Available" else x))
        ).astype("float") # convert the whole column entries to float
````

````python
# indices of entries greater than 0 and np nan in the two columns are kept

valid_row_indices = sorted((data[data["patient_survey_star_rating"] >= 0]).index.tolist() # indices of entries greater than 0 in data["patient_survey_star_rating"]
                            + (data[data["hcahps_linear_mean_value"] >= 0]).index.tolist()
                            )
````

````python
# slicing out all these valid_row_entries only
data = data.loc[valid_row_indices]
````

````python
# summing both columns up rowwise
data["master_metric"] = data[["patient_survey_star_rating", "hcahps_linear_mean_value"]].sum(1)
````

### Transformation stage
The data is reindexed to a MultiIndex of the `year` and `facility_id` column. This will give a better representation of the data by year and will be useful for the next steps

A new dataframe `finaldata` is created to make sure each hospital (`facility_id`) in a particular `year` is represented only once. This time the MultiIndex is swapped to  be `facility_id` and `year` to better represent the data to show the years a hospital occured in the the dataset.

All the unique star rating and linear score entries in `hcahps_measure_id` are represented as columns,

A list of tuples is created from the former df `data` where each tuple is for each row entry containing its `index`, its `hcahps_measure_id` value and its metric in `master_metric`. This list is used to loop through the data and insert the `master_metric` in it's right position 

```python
finaldata[index, hcahps_measure_id_value] = master_metric_value
```

````python
# a MultiIndex using the year and facility_id column for the data is created
data.set_index(["year", "facility_id"], inplace=True)
````

````python
# data.drop_duplicates(subset = ["survey_response_rate_percent", "hospital_type", "hospital_ownership", "emergency_services", "meets_criteria_for_promoting_interoperability_of_ehrs"]).shape
````

````python
# removing all duplicate facility_id entries per year and storing as a dictionary
newdict = {}
for i in (i for i in set(data.index)):
    newdict[i[-1::-1]] = data.loc[i].iloc[0].values.tolist()
````

````python
# dictionary converted to dataframe
finaldata = pd.DataFrame.from_dict(newdict, orient="index", columns= data.columns)
finaldata = finaldata.loc[sorted(finaldata.index)]
finaldata.index = pd.MultiIndex.from_tuples(finaldata.index)
finaldata = finaldata.rename_axis(index = ("facility_id", "year"))


del newdict
````

````python
# all unique facility_id entries is represented as columns
finaldata[finaldata["hcahps_measure_id"].unique()] = np.nan
````

````python
finaldata.head(20)
````

````python
# storing each master_metric entry its index and hcahps_measure_id entry
master_metric_rep = zip(((j,i) for (i,j) in data.index.tolist()), data["hcahps_measure_id"].to_list(), data["master_metric"].to_list())
````

````python
# looping through 
for i,j,k in master_metric_rep:
    finaldata.loc[i,j] = k
````

````python
# removing all other columns that are now redundant after this transformation
unuseful_col_names = ["hcahps_measure_id", "hcahps_answer_description", "patient_survey_star_rating",
                      "hcahps_answer_percent", "hcahps_linear_mean_value", "master_metric"
                      ]
finaldata.drop(columns=unuseful_col_names, inplace=True)
````

````python
finaldata.columns = [i.strip().lower().replace(" ", "_") for i in finaldata.columns]
````

````python
map = dict(data[["hcahps_measure_id", "hcahps_answer_description"]].drop_duplicates().apply(lambda x: x.str.lower().str.strip().str.replace(" ", "_")).values.tolist())
finaldata = finaldata.rename(columns = map)
finaldata = finaldata.apply(lambda x: x.replace(0, np.nan))
````
